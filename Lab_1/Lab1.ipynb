<ins>Zadanie 3</ins>:  
Zdefiniuj funkcję $ f(x)=x^2+5$ a następnie sporządź jej wykres dla zadanych przedziałów.


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): return x**2 + 5
```

```python
x1 = np.linspace(-1, 1, 100)
y1 = f(x1)

plt.plot(x1,y1)
plt.title('f(x) dla x>-1 i x<1')
plt.grid('on')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)=x^2+5'])
plt.show()
```

```python
x2 = np.linspace(-6, 6, 100)
y2 = f(x2)

plt.plot(x2,y2)
plt.title('f(x) dla x>-6 i x<6')
plt.grid('on')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)=x^2+5'])
plt.show()
```

```python
x3 = np.linspace(0, 5, 100)
y3 = f(x3)

plt.plot(x3,y3)
plt.title('f(x) dla x>-6 i x<6')
plt.grid('on')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)=x^2+5'])
plt.show()
```
