```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('../Analysis Data/5_MALOPOLSKIE.csv')
data.head()
```

### Zmienna: Dni od zakupu
Opis: Zmienna daje nam informację ile mineło dni od dokonanaj transakcji w sklepie przez danego klienta
<br>
Reszta informacji o zmiennej oraz wykres słupkowy zawierający dane są przedstawione poniżej:

```python
data['Dni od zakupu'].info()
data['Dni od zakupu'].describe(include='all')
```

```python
data1 = data.groupby('Dni od zakupu')['Marka'].count()
ax = data1.plot.bar(rot=0)
plt.ylabel('Ilość zakupień');
plt.title("Wykres słupkowy dla zmiennej 'Dni od zakupu'")
```

### Zmienna: Marka
Opis: Zmienna daje nam informację jakiej marki był zakupiony odkurzacz
<br>
Reszta informacji o zmiennej oraz wykres słupkowy zawierający dane są przedstawione poniżej:

```python
data['Marka'].info()
data['Marka'].describe(include='all')
```

```python
data2 = data.groupby('Marka')['Dni od zakupu'].count()
ax = data2.plot.bar(rot=0)
plt.ylabel('Ilość zakupień w danej firmie');
plt.title("Wykres słupkowy dla zmiennej 'Marka'")
```

### Zmienna: Wiek kupującego

### Zmienna: Ocena
